```python
li = [i for i in df.columns]
di = {'bool': [], 'float': [], 'int': [], 'object': []}

for i in li:
    if df[i].dtype == bool:
        di['bool'].append(i)
    elif df[i].dtype == float:
        di['float'].append(i)
    elif df[i].dtype == int:
        di['int'].append(i)
    elif df[i].dtype == object:
        di['object'].append(i)

for dtype, columns in di.items():
    di[dtype] = ','.join(columns)
```

```python
""" dropping columns with more than 92% values missing"""
columns_to_drop = [
    'City',
    'County',
    'DateOfBirth',
    'EmploymentPosition',
    'Rating_V0',
    'EL_V0',
    'CreditScoreEsEquifaxRisk',
    'EL_V1'
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
"""for columns that have between 25% to 92% missing values im using indicator variables to replace them"""
columns_with_high_missingness = [
    'Rating_V1',
    'Rating_V2',
    'NrOfDependants',
    'WorkExperience',
    'CreditScoreFiAsiakasTietoRiskGrade',
    'GracePeriodStart',
    'GracePeriodEnd',
    'PlannedPrincipalTillDate',
    'PreviousEarlyRepaymentsBefoleLoan',
    'ReScheduledOn',
    'NextPaymentDate',
    'EAD1',
    'EAD2',
    'PrincipalRecovery',
    'InterestRecovery',
    'PlannedInterestPostDefault',
    'PlannedPrincipalPostDefault',
    'DefaultDate',
    'ActiveLateCategory',
    'PrincipalWriteOffs',
    'InterestAndPenaltyWriteOffs',
    'PrincipalDebtServicingCost',
    'InterestAndPenaltyDebtServicingCost',
    'DebtOccuredOn',
    'CurrentDebtDaysPrimary',
    'DebtOccuredOnForSecondary',
    'CurrentDebtDaysSecondary',
    'ActiveLateLastPaymentCategory',
    'ContractEndDate',
    'CreditScoreEeMini',
    'RecoveryStage',
    'NrOfScheduledPayments',
    'NextPaymentNr',
    'PreviousRepaymentsBeforeLoan',
    'StageActiveSince',
    'WorseLateCategory'
]
for column in columns_with_high_missingness:
    df[column + '_missing'] = df[column].isnull().astype(int)

df.drop(columns=columns_with_high_missingness, inplace=True)
```

```python
"""
for columns with more than 1 missing value, first separating based on dtype

"""
# Initialize dictionaries to store column names based on data types
float_columns = []
int_columns = []
object_columns = []

# Calculate the number of missing values for each column and sort in descending order
mis = df.isnull().sum().sort_values(ascending=False)

# Iterate through the mis Series and categorize columns based on their data types
for column_name, missing_count in mis.items():
    if missing_count > 1:
        dtype = df[column_name].dtype
        # Append column names to the corresponding list based on their data types
        if dtype == 'float64':
            float_columns.append(column_name)
        elif dtype == 'int64':
            int_columns.append(column_name)
        elif dtype == 'object':
            object_columns.append(column_name)

# Print column names with more than one missing value and their corresponding data types
print("Float columns with more than one missing value:", float_columns)
print("Integer columns with more than one missing value:", int_columns)
print("Object columns with more than one missing value:", object_columns)
```

```python

# Iterate through the float columns
real_floats=[]
for column_name in float_columns:
    unique_values = df[column_name].unique()
    
    # Check if unique values are limited and discrete
    if len(unique_values) < 10 and all(isinstance(value, (int, float)) for value in unique_values):
        pass
    else:
        # If unique values are not limited and discrete, consider the column as numerical
        real_floats.append(column_name)
```

#### combining 'LoanApplicationStartedDate' and 'LoanDate' to form new column called loan wait time

```python
# Convert 'LoanApplicationStartedDate' and 'LoanDate' to datetime format
df['LoanApplicationStartedDate'] = pd.to_datetime(df['LoanApplicationStartedDate'])
df['LoanDate'] = pd.to_datetime(df['LoanDate'])

# Calculate the loan wait time
df['LoanWaitTime'] = df['LoanDate'] - df['LoanApplicationStartedDate']

# Convert the wait time to days (or any other desired unit)
df['LoanWaitTime'] = df['LoanWaitTime'].dt.days

# Display the updated DataFrame
print(df[['LoanApplicationStartedDate', 'LoanDate', 'LoanWaitTime']])
```
